```python
url = 'https://www.timesofsports.com/cricket/ipl/all-teams-logo/'

page = requests.get(url)


print(page)
print(url)
```

```python
soup = BeautifulSoup(page.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
img_box = soup.find('figure',class_='wp-caption aligncenter')
```

```python
img_box.find('img')
```

```python
img_link = img_box.find('img')['src']
print(img_link)
```

```python
img_name = img_link.split('/')[-1]
```

```python
with open(img_name,'wb') as file:

    image = requests.get(img_link)

    file.write(image.content)
```

```python
img_boxes = soup.find_all('figure',class_='wp-caption alignment')

print(len(img_boxes))
```

```python
for img_box in img_boxes:
    
    img_link = img_box.find('img')['src']
    img_name = img_link.split('/')[-1]

    with open(img_name,'wb') as file:

        image = requests.get(img_link)

        file.write(image.content)
```
